Ejercicio 16:

En un grupo de 25 alumnos, hay 13 chicas y 12 chicos


a) ¿Cuantos equipos de cinco personas puenden formarse?

- Dado que tenemos 25 personas y se deben formar equipos de 5, utilizaremos la formula de combinaciones:

$$
\begin{align}
    _nC_r &= \dfrac{n!}{r! * (n - r)!}\\
    _25C_5 &= \dfrac{25!}{5! * (25 - 5)!}\\
    _25C_5 &= \dfrac{25!}{5! * 20!} \\
    _25C_5 &= 53130
\end{align}
$$

Por lo tanto se pueden formar **53130** equipos de 5 personas

```python
# Implemeentacion de la solucion de inciso a) del ejercicio 16
n, r = 25, 5

equipos_de_5 = comb(n, r)
print('Hay ',equipos_de_5,' formas de hacer esquipos de 5 personas ')
```

Se tiene una caja con 24 latas y se sabe que dos estan contaminadas. Se selecionan tres para someterlas a pruebas de calidad

- Dado que hay 24 latas y ver cuantas combinaciones de 3 latas se pueden formar
utilizaremos la formula de combinaciones:

$$

    _nC_r &= \dfrac{n!}{r! * (n - r)!}\\

    _24C_3 &= \dfrac{24!}{3! * 21!} \\

$$

Por lo tanto hay **2024** combinaciones de tres latas
